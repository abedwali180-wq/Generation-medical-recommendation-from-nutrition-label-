```python
import pandas as pd
import json
import os

# --- إعدادات المسارات ---
EXCEL_FILE = r'C:\Users\DELL\Desktop\Recommendation_sys\CODE\Data\labeling_templateFinalVersion.xlsx' # ملف الإكسيل الذي عبأته
OUTPUT_JSONL = 'metadata.jsonl'       # الملف الناتج للنموذج

def convert_excel_to_jsonl():
    if not os.path.exists(EXCEL_FILE):
        print(f"❌ لم يتم العثور على ملف: {EXCEL_FILE}")
        return

    # 1. قراءة البيانات
    df = pd.read_excel(EXCEL_FILE)

    with open(OUTPUT_JSONL, 'w', encoding='utf-8') as f:
        for index, row in df.iterrows():
            # بناء القاموس الداخلي (فقط للقيم الموجودة)
            gt_parse = {}
            
            if pd.notna(row['saturated_fat_g']) and str(row['saturated_fat_g']).strip() != "":
                gt_parse["saturated_fat"] = str(row['saturated_fat_g'])
            
            if pd.notna(row['sodium_mg']) and str(row['sodium_mg']).strip() != "":
                gt_parse["sodium"] = str(row['sodium_mg'])
                
            if pd.notna(row['cholesterol_mg']) and str(row['cholesterol_mg']).strip() != "":
                gt_parse["cholesterol"] = str(row['cholesterol_mg'])

            # الهيكل النهائي الذي يتطلبه Donut
            # يجب أن يكون ground_truth عبارة عن JSON string داخل JSON line
            line = {
                "file_name": row['image_name'],
                "ground_truth": json.dumps({"gt_parse": gt_parse}, ensure_ascii=False)
            }
            
            # كتابة السطر في ملف JSONL
            f.write(json.dumps(line, ensure_ascii=False) + "\n")

    print(f"✅ تم بنجاح إنشاء ملف: {OUTPUT_JSONL}")
    print(f"📝 إجمالي الأسطر المعالجة: {len(df)}")

if __name__ == "__main__":
    convert_excel_to_jsonl()
```
