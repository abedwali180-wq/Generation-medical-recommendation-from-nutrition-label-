```python
import os
import json
import shutil
from sklearn.model_selection import train_test_split

# --- إعدادات المسارات (تأكد من تعديلها لمساراتك) ---
SOURCE_IMAGES_DIR = r'C:\Users\DELL\Desktop\Recommendation_sys\CODE\Data\nutrition_english_cropped' # مجلد الصور الأصلي
SOURCE_JSONL = r'C:\Users\DELL\Desktop\Recommendation_sys\CODE\Data\metadata.jsonl'  # ملف الـ JSONL الذي أنشأناه سابقاً
OUTPUT_DIR = r'C:\Users\DELL\Desktop\Recommendation_sys\CODE\Data\Donut_Dataset'     # المجلد النهائي للرفع

def split_dataset():
    # 1. إنشاء المجلدات النهائية
    train_dir = os.path.join(OUTPUT_DIR, 'train')
    val_dir = os.path.join(OUTPUT_DIR, 'validation')
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # 2. قراءة كل الأسطر من ملف JSONL الرئيسي
    with open(SOURCE_JSONL, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # 3. تقسيم الأسطر (80% تدريب، 20% تحقق)
    train_lines, val_lines = train_test_split(lines, test_size=0.20, random_state=42)

    # 4. وظيفة لمعالجة كل قسم
    def process_split(data_lines, target_folder):
        new_jsonl_path = os.path.join(target_folder, 'metadata.jsonl')
        
        with open(new_jsonl_path, 'w', encoding='utf-8') as f_out:
            for line in data_lines:
                data = json.loads(line)
                image_name = data['file_name']
                source_img_path = os.path.join(SOURCE_IMAGES_DIR, image_name)
                
                # التأكد من وجود الصورة قبل نسخها
                if os.path.exists(source_img_path):
                    # نسخ الصورة للمجلد الجديد
                    shutil.copy(source_img_path, os.path.join(target_folder, image_name))
                    # كتابة السطر في الملف الجديد
                    f_out.write(line)
                else:
                    print(f"⚠ تحذير: الصورة {image_name} غير موجودة في المصدر!")

    # 5. تنفيد العملية للقسمين
    print("🚀 بدء عملية التقسيم والنسخ...")
    process_split(train_lines, train_dir)
    print("✅ تم تجهيز مجلد التدريب (Train).")
    
    process_split(val_lines, val_dir)
    print("✅ تم تجهيز مجلد التحقق (Validation).")

    print(f"\n✨ انتهى! البيانات جاهزة في: {OUTPUT_DIR}")
    print(f"📈 التدريب: {len(train_lines)} صورة | التحقق: {len(val_lines)} صورة")

if __name__ == "__main__":
    split_dataset()
```
